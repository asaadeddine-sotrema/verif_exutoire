```python
import pandas as pd

def get_header_row(file_path, keyword):
    try:
        df_temp = pd.read_excel(file_path, header=None, nrows=20)
        for idx, row in df_temp.iterrows():
            line_text = " ".join(row.astype(str).values).lower()
            if keyword.lower() in line_text:
                return idx
        return 0
    except Exception:
        return 0

# --- Chargement ---
df_valene = pd.read_excel('source/Export Sotrema novembre 2025.xlsx', header=get_header_row('source/Export Sotrema novembre 2025.xlsx', 'N° de pesée'))
df_pap = pd.read_excel('source/VALENE PAP Novembre 2025.xlsx', header=get_header_row('source/VALENE PAP Novembre 2025.xlsx', 'Description'))
df_pav = pd.read_excel('source/VALENE PAV Novembre 2025.xls', header=get_header_row('source/VALENE PAV Novembre 2025.xls', 'Description'))
df_sotrema2 = pd.read_excel('source/VALENE SOTREMA2 Novembre 2025.xlsx', header=get_header_row('source/VALENE SOTREMA2 Novembre 2025.xlsx', 'Description'))
```

```python
df_valene = df_valene.rename(columns={'N° de pesée': 'Num Ticket', 'Poids de la matière': 'Poids en Tonnes'})
```

```python
df = pd.merge(df_kerlog, df_valene, on='Num Ticket', suffixes=('_kerlog', '_valene'), how='inner')
df['Num Bon'] = df['Num Bon'].astype(int)
df['Date d\'entrée'] = pd.to_datetime(df['Date d\'entrée']).dt.date

print("Aperçu des données fusionnées (premières lignes) :")
display(df.head())
```

```python
df['Difference'] = df['Poids en Tonnes_kerlog'] - df['Poids en Tonnes_valene']
ecarts = df[df['Difference'].abs() > 0]

if not ecarts.empty:
    print(f"\nDes écarts significatifs ont été trouvés pour {len(ecarts)} enregistrements :")
    display(ecarts)
else:
    print("\nAucun écart significatif trouvé dans les 'Poids en Tonnes' entre df_kerlog et df_valene pour les 'Num Ticket' correspondants.")

print(f"\nNombre total d'enregistrements avec un 'Num Ticket' correspondant : {len(df)}")

print(f"\nLe taux de fiabilisation des enregistrements est de {round(1-(len(ecarts)/len(df)), 2)} %.")
```

```python
num_ticket_valene = set(df_valene['Num Ticket'].dropna())
num_ticket_kerlog = set(df['Num Ticket'].dropna())

num_ticket_manquant = num_ticket_valene - num_ticket_kerlog
num_ticket_manquant

len(num_ticket_manquant)

df.to_csv('verif_tonnage_resultat.csv', index=False)
```

```python
df_bons_manquants = df_valene[df_valene['Num Ticket'].isin(num_ticket_manquant)].copy()

colonnes_a_afficher = ['Num Ticket', "Date d'entrée"]

cols_existantes = [c for c in colonnes_a_afficher if c in df_bons_manquants.columns]

df_bons_manquants['Date d\'entrée'] = pd.to_datetime(df_bons_manquants['Date d\'entrée']).dt.date

display(df_bons_manquants[cols_existantes])
```
